```python
# 2.2.2

import osmium as osm
import requests
import folium
from folium.plugins import MarkerCluster, HeatMap
```

```python
# 2.2.2 obtaining data

url = 'https://download.geofabrik.de/europe/netherlands/groningen-latest.osm.pbf'

response = requests.get(url, stream=True)

filepath = 'groningen-latest.osm.pbf'
with open(filepath, 'wb') as file:
    for chunk in response.iter_content(chunk_size=1024):
        if chunk:
            file.write(chunk)


```

```python
# 2.2.2 create bench handler

class BenchHandler(osm.SimpleHandler):
    def __init__(self):
        osm.SimpleHandler.__init__(self)
        self.benches = []

    def node(self, n):
        if 'amenity' in n.tags and n.tags['amenity'] == 'bench':
            self.benches.append((n.location.lat, n.location.lon))

handler = BenchHandler()

handler.apply_file('groningen-latest.osm.pbf')

print(f'Found {len(handler.benches)} benches.')
print(f'Example bench: {handler.benches[0]}')
```

```python
# 2.2.2 Plot benches and heatmap

center_loc = handler.benches[0]
benches_groningen = folium.Map(location = center_loc, zoom_start = 12)
benches_groningen.save('benches_groningen.html')

marker_cluster = MarkerCluster().add_to(benches_groningen)

for lat, lon in handler.benches:
    folium.CircleMarker(
        location = [lat, lon],
        radius = 5,
        color = 'blue',
        fill=True,
        fill_color = 'blue',
        fill_opacity = 0.2

    ).add_to(marker_cluster)

benches_groningen.save('benches_groningen.html')

benches_groningen

HeatMap(handler.benches, radius = 15, blur = 10).add_to(benches_groningen)

benches_groningen.save('benches_groningen.html')

benches_groningen

```
