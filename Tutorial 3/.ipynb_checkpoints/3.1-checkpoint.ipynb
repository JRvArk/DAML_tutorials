```python
import pandas as pd
import numpy as np
from sklearn.linear_model import RidgeCV, LassoCV, Ridge, Lasso
from sklearn.model_selection import train_test_split, KFold
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import mean_absolute_percentage_error
```

# 3.1

```python
# 3.1 Read data
df = pd.read_csv("../ToyotaCorolla.csv")
df = df.drop(columns=['Id', 'Model', 'Mfg_Month', 'Mfg_Year', 'Cylinders'])
df['Fuel_Type'] = df['Fuel_Type'].astype('category')
df['Color'] = df['Color'].astype('category')
# df['Gears'] = df['Gears'].astype('category')
# df['Doors'] = df['Doors'].astype('category')

df.loc[df['CC'] == 16000, 'CC'] = 1600
df.rename(columns={'Age_08_04': 'Age'}, inplace=True)

for col in df.columns:
    if df[col].nunique() == 2:
        df[col] = df[col].astype('bool')

df['Combined_Color'] = df.apply(
    lambda row:
    f'Met_{row["Color"]}' if row['Met_Color']
    else row['Color'], axis=1
).astype('category')

df['Age^2'] = df['Age']**2
df['KM^2'] = df['KM']**2
df['eAge'] = np.exp(-df['Age'])
df['eKM'] = np.exp(-df['KM'])
df['Weight^2'] = df['Weight'] ** 2

df.describe(percentiles=[.5]).T
```

```python
df.select_dtypes(bool).apply(np.mean)
```

```python
# 3.1 Make design matrix and target vector, and split
#features = ['KM', 'Weight', 'Age', 'HP', 'CC', 'Quarterly_Tax', 'Guarantee_Period', 'ABS', 'Airco', 'Boardcomputer', 'Fuel_Type_Diesel', 'Fuel_Type_Petrol']
features = ['Sport_Model', 'Fuel_Type', 'CC', 'Gears', 'Weight', 'HP', 'KM', 'Age', 'ABS', 'Airco']
#           'Quarterly_Tax', 'Guarantee_Period']
extras = ['Automatic', 'ABS', 'Airbag_1', 'Airbag_2', 'Airco', 'Automatic_airco', 'Boardcomputer', 'CD_Player', 'Central_Lock',
          'Powered_Windows', 'Power_Steering', 'Radio', 'Mistlamps', 'Backseat_Divider', 'Metallic_Rim', 'Radio_cassette', 
          'Parking_Assistant', 'Tow_Bar', 'Combined_Color', 'Mfr_Guarantee', 'BOVAG_Guarantee']
#features.extend(extras)
X = df[features]
y = df['Price']

cat_features = list(set(df.select_dtypes('category').columns).intersection(set(features)))
num_features = list(set(df.select_dtypes('number').columns).intersection(set(features)))
bool_features = list(set(df.select_dtypes('bool').columns).intersection(set(features)))

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, random_state=42, test_size=0.2)

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_features),
    ('cat', OneHotEncoder(drop = 'first'), cat_features),
    ], remainder='passthrough'
)

ridge_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('ridge', Ridge())
])

alphas = np.logspace(-3, 3, 100)
coefficients = []

for alpha in alphas:
    ridge_pipeline.set_params(ridge__alpha=alpha)
    ridge_pipeline.fit(X_train, y_train)
    coefficients.append(ridge_pipeline.named_steps['ridge'].coef_)

coefficients = np.array(coefficients)

# Chat gpt
# Get the numeric feature names after preprocessing
# num_feature_names = [f'num__{name}' for name in num_features]

# # Identify the indices for the numeric features
# num_indices = [i for i, feature_name in enumerate(ridge_pipeline.named_steps['preprocessor'].get_feature_names_out(X.columns))
#                if feature_name in num_feature_names]

# plt.figure(figsize=(10,10))
# for i in num_indices:
#     plt.plot(alphas, coefficients[:,i], label=num_feature_names[i]) #num_indices.index(i))
plt.figure(figsize=(10,10))
for i in range(coefficients.shape[1]):
    plt.plot(alphas, coefficients[:,i], label=ridge_pipeline[0].get_feature_names_out(X_train.columns)[i]) 
    
plt.xscale('log')
plt.xlabel('Alpha (log scale)')
plt.ylabel('Feature Weight')
plt.title('Ridge regression: feature weights vs alpha')
plt.tight_layout()
plt.legend()
plt.show()
```

##### From the plot above, we can learn that most features shrink in weight size with increasing alpha, some of them to 0, as expected.

```python
# 3.1 Make the same plot for Lasso

lasso_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('lasso', Lasso(max_iter=10_000, tol=0.001))
])

alphas = np.logspace(-2.5, 2.5, 100)
coefficients = []

for alpha in alphas:
    lasso_pipeline.set_params(lasso__alpha=alpha)
    lasso_pipeline.fit(X_train, y_train)
    coefficients.append(lasso_pipeline.named_steps['lasso'].coef_)

coefficients = np.array(coefficients)
```

```python
plt.figure(figsize=(10,10))
for i in range(coefficients.shape[1]):
    plt.plot(alphas, coefficients[:,i], label=lasso_pipeline[0].get_feature_names_out(X_train.columns)[i]) 
    
plt.xscale('log')
plt.xlabel('Alpha (log scale)')
plt.ylabel('Feature Weight')
plt.title('Lasso reg: feature weights vs alpha')
plt.tight_layout()
plt.legend()
plt.show();
```

```python
# 3.1 Doing lasso cv including all variables
features = ['Sport_Model', 'Fuel_Type', 'CC', 'Gears', 'Weight', 'HP', 'KM', 'KM^2', 'eKM', 'Age', 'Age^2', 'eAge',
            'Quarterly_Tax', 'Guarantee_Period']
extras = ['Automatic', 'ABS', 'Airbag_1', 'Airbag_2', 'Airco', 'Automatic_airco', 'Boardcomputer', 'CD_Player', 'Central_Lock',
          'Powered_Windows', 'Power_Steering', 'Radio', 'Mistlamps', 'Backseat_Divider', 'Metallic_Rim', 'Radio_cassette', 
          'Parking_Assistant', 'Tow_Bar', 'Combined_Color', 'Mfr_Guarantee', 'BOVAG_Guarantee']
features.extend(extras)
X = df[features]
y = df['Price']

cat_features = list(set(df.select_dtypes('category').columns).intersection(set(features)))
num_features = list(set(df.select_dtypes('number').columns).intersection(set(features)))
bool_features = list(set(df.select_dtypes('bool').columns).intersection(set(features)))

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, random_state=2, test_size=0.2)

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_features),
    ('cat', OneHotEncoder(drop = 'first'), cat_features),
    ], remainder='passthrough'
)

cv = KFold(n_splits=5) #, shuffle=True, random_state=0)
lassoCV_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('lassoCV', LassoCV(alphas = np.logspace(-4,4,100), cv=cv, max_iter=10_000, tol=0.001))
])

lassoCV_pipeline.fit(X_train, y_train)

mse_path = lassoCV_pipeline.named_steps['lassoCV'].mse_path_
alphas = lassoCV_pipeline.named_steps['lassoCV'].alphas_
```

```python
fold_vars = []
for _, test_idx in cv.split(X_train, y_train):
    y_test_fold = y_train[y_train.index[test_idx]]
    fold_vars.append(y_test_fold.var(ddof=1))
fold_vars = np.array(fold_vars)
scores = 1 - mse_path / fold_vars
lasso_results = pd.DataFrame(scores, index=alphas)
lasso_results
```

```python
lasso_results_means = lasso_results.apply(np.mean, axis=1)
lasso_results_means
```

```python
best_alpha = lasso_results_means.idxmax()
best_score = lasso_results_means.max()

print(f"Best Alpha: {best_alpha}\nBest Score: {best_score}")
```
