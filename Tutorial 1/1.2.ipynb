```python
# 1.2.2

import sqlite3
import pandas as pd

db_path = '/Users/user/Documents/GitHub/DAML_tutorials/Tutorial 1/waste_for_students.db'

conn = sqlite3.connect(db_path)

query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(query, conn)

print("Tables in the database:")
print(tables)

table1 = "arrivals"
table2 = "locations"

arrivalsDF= pd.read_sql(f"SELECT * FROM {table1}", conn)
locationsDF = pd.read_sql(f"SELECT * FROM {table2}", conn)

print(arrivalsDF.head(10))
print(locationsDF.head(10))

conn.close()

import folium
from folium.plugins import MarkerCluster

center_coords = [53.2194, 6.5665] 

map_groningen = folium.Map(location=center_coords, zoom_start=12)

map_groningen.save("groningen_map.html")

map_groningen

marker_cluster = MarkerCluster().add_to(map_groningen)

for index, row in locationsDF.iterrows():
    #popUpText = f"<div style=\"font-size:10px;\">Location ID: {row['name']}</div>"
    tooltip_text = f"Location {row['name']}"
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        #popup=folium.Popup(popUpText, max_width=50), 
        radius=5,  
        color="blue",  
        fill=True,
        fill_color="blue",  
        fill_opacity=0.7,
        tooltip = tooltip_text
    ).add_to(marker_cluster)

map_groningen.save("groningen_map.html")

map_groningen
```
